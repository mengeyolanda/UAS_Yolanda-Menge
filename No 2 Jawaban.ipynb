```python
#a.Apabila Cuaca buruk dengan nilai = 1, Weekday, dan Game = 0, maka berapa roti yang harus dibuat?
#  misalkan hari misterius = H-M (Weekday)
data = np.array([[5.,3.,'Weather V-1'],[1.,4.,'Weather V-1'],[4.,2.,'Weather V-1'],[1.,1.,'Holiday V-2'],[1.,1.,'Holiday V-2'],[0.,0.,'Holiday V-2'],[0.,1.,'Game V-3'],[0.,1.,'Game V-3'],[0.,0.,'Game V-3'],[1.,0.,'Weekday H-M']])
query = [1.,0.,'Weekday H-M']
```

```python
df = pd.DataFrame(data)
df.columns = ['x','y','Qty']
df
```

```python
for i in range(10):
    if(df.iloc[i]['Qty'] == 'Weather V-1'):
        plt.scatter(df.iloc[i]['x'], df.iloc[i]['y'], s=100, c='r')
    elif(df.iloc[i]['Qty'] == 'Weekday H-M'):
        plt.scatter(df.iloc[i]['x'], df.iloc[i]['y'], s=200, c='y')
    else:
     plt.scatter(df.iloc[i]['x'], df.iloc[i]['y'], s=100, c='b')
       
        
plt.grid()
plt.show()
```

```python
import math
dis = []
for i in range(10):
    dis.append(math.sqrt((float(df.iloc[i]['x']) - query[1]) **2 + (float(df.iloc[i]['y']) - query[0]) **2))
```

```python
df['dis'] = dis
df
```

```python
df.sort_values('dis')
```

```python
df.to_excel('C:/yolanda/datamining/outputNo2(a).xls')
```
